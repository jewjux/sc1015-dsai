```python
import pandas as pd

df = pd.read_csv('cleaned_1015.csv')

df
```

```python
df = df.dropna(subset=['KEYWORDS'])

df["KEYWORDS"] = df["KEYWORDS"].apply(eval)

for i, l in enumerate(df["KEYWORDS"]):
 print("list",i,"is",type(l))
```

```python
df
```

```python
def to_1D(series):
 return pd.Series([x for _list in series for x in _list])

keywordsdf = to_1D(df['KEYWORDS'])

keywordsdict = dict(keywordsdf.value_counts())
```

```python
print(len(keywordsdict))

for k in list(keywordsdict.keys()):
    if keywordsdict[k] < 10:
        del keywordsdict[k]

print(len(keywordsdict))
```

```python
pd.DataFrame.from_dict(keywordsdict.items())
```

```python
for i in range(len(df)):
    keywords_list = df.iloc[i]['KEYWORDS']
    for kw in keywords_list[:]:
        if kw not in keywordsdict:
            keywords_list.remove(kw)
```

```python
count=0
print("df length = ", len(df))
for i in range(len(df)):
    keywords_list = df.iloc[i]['KEYWORDS']
    if keywords_list == []:
        count+=1

count
```

```python
sorted_values = sorted(keywordsdict.values(), reverse=True)
top_10 = sorted_values[:20]

new_kwdict = {}
for key, value in keywordsdict.items():
    if value in top_10:
        new_kwdict[key] = value

new_kwdict
```

```python
for i in range(len(df)):
    keywords_list = df.iloc[i]['KEYWORDS']
    for kw in keywords_list[:]:
        if kw not in new_kwdict:
            keywords_list.remove(kw)

count=0
droplist = []
print("df length = ", len(df))
for i in range(len(df)):
    keywords_list = df.iloc[i]['KEYWORDS']
    if not keywords_list:
        droplist.append(i)
        count+=1

df = df.drop(index=droplist)

print("empty kw list =", count)
print("df length = ", len(df))
df
```
